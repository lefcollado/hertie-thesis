#### Transform motivo_o and motivo_d into categorical variable with labels

```python
motivo_labels = {
    1: "Trabalho Indústria",
    2: "Trabalho Comércio",
    3: "Trabalho Serviços",
    4: "Escola/Educação",
    5: "Compras",
    6: "Médico/Dentista/Saúde",
    7: "Recreação/Visitas/Lazer",
    8: "Residência",
    9: "Procurar Emprego",
    10: "Assuntos Pessoais",
    11: "Refeição"
}

pod_17_2["motivo_o_cat"] = pd.Categorical(
    pod_17_2["motivo_o"].map(motivo_labels),
    categories=list(motivo_labels.values())
)
```

```python
pod_17_2["motivo_o_cat"].value_counts(dropna=False)
```

```python
pod_17_2["motivo_d_cat"] = pd.Categorical(
    pod_17_2["motivo_d"].map(motivo_labels),
    categories=list(motivo_labels.values())
)
```

```python
pod_17_2["motivo_d_cat"].value_counts(dropna=False)
```

#### Grouping categories

```python
pod_17_2["motivo_o_group"] = pod_17_2["motivo_o_cat"].map({
    "Residência": "Home",
    
    "Trabalho Serviços": "Work",
    "Trabalho Comércio": "Work",
    "Trabalho Indústria": "Work",
    
    "Escola/Educação": "Education",
    
    "Assuntos Pessoais": "Personal Matters",
    "Compras": "Personal Matters",
    "Refeição": "Personal Matters",
    "Procurar Emprego": "Personal Matters",
    "Recreação/Visitas/Lazer": "Personal Matters",
    "Médico/Dentista/Saúde": "Personal Matters"
})
```

```python
pod_17_2["motivo_o_group"] = pd.Categorical(
    pod_17_2["motivo_o_group"],
    categories=["Home", "Work", "Education", "Personal Matters"]
)
```

```python
pod_17_2["motivo_o_group"].value_counts(dropna=False)
```

```python
pod_17_2["motivo_d_group"] = pod_17_2["motivo_d_cat"].map({
    "Residência": "Home",
    
    "Trabalho Serviços": "Work",
    "Trabalho Comércio": "Work",
    "Trabalho Indústria": "Work",
    
    "Escola/Educação": "Education",
    
    "Assuntos Pessoais": "Personal Matters",
    "Compras": "Personal Matters",
    "Refeição": "Personal Matters",
    "Procurar Emprego": "Personal Matters",
    "Recreação/Visitas/Lazer": "Personal Matters",
    "Médico/Dentista/Saúde": "Personal Matters"
})
```

```python
pod_17_2["motivo_d_group"] = pd.Categorical(
    pod_17_2["motivo_d_group"],
    categories=["Home", "Work", "Education", "Personal Matters"]
)
```

```python
pod_17_2["motivo_d_group"].value_counts(dropna=False)
```

```python
pod_17_2["motivo_d_group"].dtype
```

#### Join the two columns

```python
pod_17_2["motivo"] = pod_17_2.apply(
    lambda row:
        "Home - Work"
        if {str(row["motivo_o_group"]), str(row["motivo_d_group"])} == {"Home", "Work"}
        else "Home - Personal Matters"
        if {str(row["motivo_o_group"]), str(row["motivo_d_group"])} == {"Home", "Personal Matters"}
        else "Home - Education"
        if {str(row["motivo_o_group"]), str(row["motivo_d_group"])} == {"Home", "Education"}
        else "Others",
    axis=1
)

pod_17_2["motivo"].value_counts(dropna=False)
```

### Level of education

```python
grau_labels = {
    1: "No Formal Education/Incomplete Primary Education",
    2: "Primary Education Completed / Lower Secondary Incomplete",
    3: "Lower Secondary Completed / Upper Secondary Incomplete",
    4: "Upper Secondary Completed / Higher Education Incomplete",
    5: "Higher Education Completed"
}

pod_17_2["grau_ins"] = pd.Categorical(
    pod_17_2["grau_ins"].map(grau_labels),
    categories=list(grau_labels.values()),
    ordered=True
)
```

```python
pod_17_2["h_saida_period"] = pd.cut(
    pod_17_2["h_saida"],
    bins=[0, 5, 10, 13, 18, 24],
    labels=["Dawn", "Morning", "Lunch", "Afternoon", "Night"],
    include_lowest=True,
    right=True
)

pod_17_2["h_saida_period"].value_counts(dropna=False)
```

### In which sector is your main occupation?

```python
pod_17_2["setor1"].value_counts(dropna=False)
```

```python
sector_map = {
    1: "Agriculture",
    2: "Construction",
    3: "Industry",
    4: "Commerce",
    13: "Public Administration"
}

pod_17_2["sector_comb"] = pod_17_2["setor1"].map(sector_map)

pod_17_2["sector_comb"] = pod_17_2["sector_comb"].mask(
    pod_17_2["sector_comb"].isna() & pod_17_2["setor1"].notna(),
    "Services"
)

pod_17_2["sector_comb"].value_counts(dropna=False)
```

### Emplyment relationship

```python
pod_17_2["vinc1"].value_counts(dropna=False)
```

```python
empl_map = {
    1: "Wage and salaried employees",
    2: "Wage and salaried employees",
    3: "Public sector",
    4: "Self-employed"
}

pod_17_2["empl_relat"] = pod_17_2["vinc1"].map(empl_map)

pod_17_2["empl_relat"] = pod_17_2["empl_relat"].mask(
    pod_17_2["empl_relat"].isna() & pod_17_2["vinc1"].notna(),
    "Other"
)

pod_17_2["empl_relat"].value_counts(dropna=False)
```

## Creating new dataframe with constructed variables

```python
summary = pd.DataFrame({
    "column": pod_17_2.columns,
    "dtype": pod_17_2.dtypes.values,
    "na_count": pod_17_2.isna().sum().values
})

# Function to get categories / unique values (limited for readability)
def get_values(col):
    if isinstance(col.dtype, pd.CategoricalDtype):
        return ", ".join(map(str, col.cat.categories))
    elif col.nunique() <= 10:
        return ", ".join(map(str, col.dropna().unique()))
    else:
        return "—"

summary["possible_values"] = [
    get_values(pod_17_2[col]) for col in pod_17_2.columns
]

summary.style.set_properties(**{"text-align": "left"})
```
